```python
# 📌 Step 0: Install Required Libraries
!pip install transformers PyPDF2 pandas --quiet

# 📌 Step 1: Upload ESG Report (PDF)
from google.colab import files
print("📄 Upload your ESG Report (PDF format only)")
pdf_uploaded = files.upload()

from PyPDF2 import PdfReader
pdf_filename = next((f for f in pdf_uploaded if f.endswith(".pdf")), None)

if pdf_filename:
    reader = PdfReader(pdf_filename)
    text = "".join(page.extract_text() or "" for page in reader.pages)
    print("✔️ PDF text extracted.")
else:
    print("❌ No valid PDF found.")
    text = None

# 📌 Step 2: Upload ESG Criteria (Excel)
print("\n📊 Upload your ESG criteria checklist (Excel format only)")
excel_uploaded = files.upload()

import pandas as pd
excel_filename = next((f for f in excel_uploaded if f.endswith((".xlsx", ".xls"))), None)

if excel_filename:
    df = pd.read_excel(excel_filename)
    esg_criteria = df.iloc[:, 0].dropna().tolist()
    print(f"✔️ {len(esg_criteria)} ESG criteria loaded.")
else:
    print("❌ No valid Excel file found.")
    esg_criteria = None

# 📌 Step 3: Load Hugging Face Model
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch

print("\n⏳ Loading Hugging Face model...")
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-large")
model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-large")
print("✅ Model loaded.")

# 📌 Step 4: ESG Disclosure Analysis
def scan_with_huggingface(text, criteria_list):
    prompt = f"""
You are an ESG analyst. Review the following ESG report content and assess the level of disclosure for each criterion listed.

Report Content:
{text[:2000]}

Criteria:
{', '.join(criteria_list[:10])}

Respond with:
- Disclosure Status: Fully Disclosed / Partially Disclosed / Not Disclosed
- Suggested Improvement (if any)

Use simple bullet points.
"""
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True)
    outputs = model.generate(**inputs, max_length=1024)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response

# 📌 Step 5: Run if Inputs Are Valid
if text and esg_criteria:
    print("\n📋 ESG Disclosure Analysis:\n")
    print(scan_with_huggingface(text, esg_criteria))
else:
    print("⚠️ Skipping analysis: Missing report text or ESG criteria.")
```
